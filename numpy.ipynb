```python
import numpy as np

a=np.array(6)
b=np.array([2,6,8,5])
c=np.array([[2,4,4],[6,6,7]])
d=np.array([[[4,6,7],[4,6,8]],[[5,7,8],[6,7,8]]])
print("zero dimentional:",a)
print("1-D:",b)
print("2-D:",c)
print("3-D:",d)

print(a.ndim)
print(b.ndim)
print(c.ndim)
print(d.ndim)
```

```python
import matplotlib.pyplot as plt
x=np.array([0,4])
y=np.array([0,4])
plt.plot(x,y)
plt.show
```

```python
import matplotlib.pyplot as pyplot

x=np.array([0,1,2,3,4])
y=np.array([0,1,2,3,4])
plt.plot(x,y, marker="*")
plt.title("Wanjiru")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.grid()
plt.show()
```

```python
x_axis=np.array([0,3,5,7])
y_axis=np.array([0,5,8,16])
plt.scatter(x_axis,y_axis)
plt.show()
```

```python
x=np.array(["Apples","Bananas","mango","grapes"])
y=np.array([4,15,20,2])
plt.bar(x,y,color="green",width=0.4)
plt.show()
```

```python
x=np.array(["Apples","Bananas","mango","grapes"])
y=np.array([4,15,20,2])
plt.barh(x,y, height=0.5,color="red")
plt.show()
```

```python
x=np.array([70,50,10,90])
my_label=["BMW","Jeep","Toyota","Volvo"]
my_colors=("green","violet","black","yellow")
my_explode=[0.3,0,0,0]
plt.pie(x,labels=my_label, colors=my_colors,explode=my_explode,shadow=True)
plt.legend(title="cars")
plt.show()
```
